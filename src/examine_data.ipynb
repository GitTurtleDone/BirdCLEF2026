```python
from pathlib import Path
import soundfile as sf
files = list(Path('../data/train_audio').rglob("*.ogg"))
print(f'Number of audio files in train_audio: {len(files)}')
```

#### Check audio duration distribution

```python
import matplotlib.pyplot as plt
import numpy as np
durations = [sf.info(str(file)).duration for file in files]
print(f'Average audio duration: {sum(durations)/len(durations):.2f} seconds')

plt.hist(durations, bins=100, edgecolor='black')
plt.title('Distribution of Audio Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Frequency')
plt.yscale('log')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

### Check species distribution

```python
species = '1161364'
taxonomy_df[taxonomy_df['primary_label'] ==  species]
```

```python
species_counts = train_df['primary_label'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(range(len(species_counts)), species_counts)
plt.xlabel('Species')
plt.ylabel('Counts')
plt.title('Distribution of Species in Train Dataset')
plt.tight_layout()
plt.show()

print(f'Min samples: {species_counts.min()} \n \
Max samples: {species_counts.max()} \n \
Mean samples: {species_counts.mean():.2f} \n \
Median samples: {species_counts.median():.2f}')
```

```python
class_counts = train_df['class_name'].value_counts()
plt.bar(range(len(class_counts)), class_counts)
plt.xlabel('Class')
plt.ylabel('Counts')
plt.yscale('log')    
plt.title('Distribution of classes in the train dataset')
plt.tight_layout()
plt.show()
print(f'Min class counts: {class_counts.min()}')
print(f'Max class counts: {class_counts.max()}')
print(f'Mean class counts: {class_counts.mean():.2f}')
print(f'Meidan class counts: {class_counts.median()}')
```
